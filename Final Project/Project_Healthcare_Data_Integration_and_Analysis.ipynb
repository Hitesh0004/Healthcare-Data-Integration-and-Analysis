### 3.1a. Write the Python code to identify the Number of missing values.

### 3.1c Write the Python code to identify inconsistent categorical entries in the Gender column.

### 3.1d Write the Python code to identify inconsistent date formats in the AdmissionDate column.

```python
# View unique AdmissionDate values
print("Sample AdmissionDate values:")
print(df["AdmissionDate"].head(20))

# Try converting AdmissionDate to datetime
converted_dates = pd.to_datetime(df["AdmissionDate"], errors="coerce", infer_datetime_format=True)

# Count dates that could not be converted
invalid_dates = converted_dates.isnull().sum()

print("Number of inconsistent or invalid AdmissionDate values:")
print(invalid_dates)

# Display rows with invalid AdmissionDate values
df[converted_dates.isnull()]
```

### 3.2a Write the Python code to handle missing values appropriately.

```python
# 3.2a Handle missing values appropriately

# Fill missing DateOfBirth values with 'Unknown'
df["DateOfBirth"] = df["DateOfBirth"].fillna("Unknown")

# Fill missing ZipCode values with 'Unknown'
df["ZipCode"] = df["ZipCode"].fillna("Unknown")

# Check missing values after handling
print("Missing values after cleaning:")
print(df.isnull().sum())
```

### 3.2b Write the Python code to remove exact duplicate rows.

```python
# 3.2b Identify and remove exact duplicate rows, keeping only the first occurrence

# Count duplicate rows before cleaning
print("Number of duplicate rows before cleaning:")
print(df.duplicated().sum())

# Display all duplicate records, including the first occurrence
duplicate_records = df[df.duplicated(keep=False)]

print("\nDuplicate records:")
display(duplicate_records)

# Remove exact duplicate rows, keeping only the first occurrence
df = df.drop_duplicates(keep="first")

# Check duplicates after removal
print("\nNumber of duplicate rows after cleaning:")
print(df.duplicated().sum())
```

### 3.2c Write the Python code to standardize gender values.

```python
# 3.2c Standardize Gender values

df["Gender"] = df["Gender"].astype(str).str.strip().str.lower()

df["Gender"] = df["Gender"].replace({
    "m": "Male",
    "male": "Male",
    "f": "Female",
    "female": "Female"
})

# Check Gender values after standardisation
print("Gender values after standardisation:")
print(df["Gender"].value_counts())
```

### 3.2d Write the Python code to standardize date values.

```python
# 3.2d Standardize AdmissionDate values without creating an original column

# Clean AdmissionDate as text
date_col = df["AdmissionDate"].astype(str).str.strip()

# Convert possible Excel serial dates to numeric
numeric_dates = pd.to_numeric(date_col, errors="coerce")

# Identify Excel serial dates such as 45258 or 45439
is_excel_serial = numeric_dates.notna()

# Create empty datetime series
standardized_dates = pd.Series(pd.NaT, index=df.index)

# Convert Excel serial dates
standardized_dates.loc[is_excel_serial] = pd.to_datetime(
    numeric_dates.loc[is_excel_serial],
    unit="D",
    origin="1899-12-30",
    errors="coerce"
)

# Convert normal date strings, e.g. 07-06-2024
standardized_dates.loc[~is_excel_serial] = pd.to_datetime(
    date_col.loc[~is_excel_serial],
    dayfirst=True,
    errors="coerce"
)

# Save standardized dates back to AdmissionDate
df["AdmissionDate"] = standardized_dates.dt.strftime("%Y-%m-%d")

# Check results
print("AdmissionDate after standardisation:")
print(df[["PatientID", "AdmissionDate"]].head(20))

print("\nInvalid or missing AdmissionDate values after conversion:")
print(df["AdmissionDate"].isnull().sum())
```

```python
# Check duplicate rows again

duplicate_count = df.duplicated().sum()

print("Number of duplicate rows:")
print(duplicate_count)

# Display duplicate records if any
duplicate_records = df[df.duplicated(keep=False)]

print("\nDuplicate records:")
display(duplicate_records)
```

### 4.1a Write an SQL query to list all patients born before 1980.  

```python
run_query("""
SELECT *
FROM patients
WHERE DATE(DateOfBirth) < '1980-01-01';
""")
```

### 4.1b Write an SQL query to list all female patients with Type 2 Diabetes.

```python
run_query("""
SELECT
   *
FROM patients
WHERE Gender = 'Female'
  AND PrimaryCondition LIKE '%Type 2 Diabetes%';
""")
```

### 4.1c Write an SQL query to display all lab tests performed. List each patient (PatientID, Gender, and DateOfBirth from *patients*) with their Lab test name and result value (TestName and TestResultValue from *lab_results*).

```python
run_query("""
SELECT
    p.PatientID,
    p.Gender,
    p.DateOfBirth,
    l.TestName,
    l.TestResultValue
FROM patients p
JOIN lab_results l
    ON p.PatientID = l.PatientID;
""")
```

### 4.1d Write an SQL query to display all distinct patients with their payer(s). List each patient (PatientID, Gender, and DateofBirth from *patients*) with their payer(s) from *pharmacy_claims*).

```python
run_query("""
SELECT DISTINCT
    p.PatientID,
    p.Gender,
    p.DateOfBirth,
    pc.Payer
FROM patients p
JOIN pharmacy_claims pc
    ON p.PatientID = pc.PatientID;
""")
```
